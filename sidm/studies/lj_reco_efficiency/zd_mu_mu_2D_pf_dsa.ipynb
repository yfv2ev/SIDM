- We see unexpected behavior occurs specifically in the DSA-DSA case.
- We see low efficiency for the DSA-DSA plots for high $\Delta$R.
- From the $\Delta$R vs $Z_d$ $p_T$ plot, for a fixed $\Delta$R, we see high efficiency for lower $p_T$ regions. For a fixed $p_T$, we see high efficiency for low $\Delta$R 
- From the $Z_d$ $p_T$ vs $Z_d$ $L_{xy}$ plots , we see low efficiency for lower $p_T$ and high $L_{xy}$ region. For fixed $p_T$, low $L_{xy}$ has higher efficiency.
- We see similar behaviour for $\Delta$R vs Leading/Subleading Muons $p_T$
