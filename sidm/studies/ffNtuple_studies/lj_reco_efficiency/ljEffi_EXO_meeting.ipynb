### This is a notebook created to present the lepton jet reconstruction efficiency to the EXO Long lived meeting. Here we are presenting the LJ -reconstruction efficiency w.r.t Dark Photon Pt and Lxy. Here we consider two samples with same bound state mass (200 Gev) and avg transverse decay length (300cm). Then we consider two dark photon masses 0.25 GeV (most collimated) and 5 GeV (least collimated).

```python
samples = [
    #"2Mu2E_100GeV_5GeV_400mm",
    #"2Mu2E_100GeV_5GeV_400mm",
    #"2Mu2E_150GeV_5GeV_2p7mm",
    "2Mu2E_200GeV_0p25GeV_10mm",
     "2Mu2E_200GeV_5GeV_200mm",
    #"2Mu2E_500GeV_5GeV_80mm",
    #"2Mu2E_800GeV_5GeV_0p5mm",
    #"2Mu2E_1000GeV_5GeV_0p4mm",
    #"2Mu2E_1000GeV_5GeV_0p04mm",
    #"2Mu2E_1000GeV_5GeV_20mm",
    #"2Mu2E_100GeV_1p2GeV_9p6mm",
    #"2Mu2E_1000GeV_5GeV_40mm",
    #"2Mu2E_100GeV_p25GeV_400mm",
   # "2Mu2E_1000GeV_5GeV_4mm",
    
]
fileset = utilities.make_fileset(samples, "ffntuple_v4")
```

The selections I consider are the following : <br>
1) baseNoLj - Includes all the events <br>
2) baseNoLj_lowLxyA - Includes events with $Z_d$ $L_{xy}$ < 150 cm<br>
3) baseNolj_highPtA - Includes events with $Z_d$ $p_T$ >30 GeV<br>
4) baseNoLj_Lxy_upto250A - Includes events with $Z_d$ $L_{xy}$ < 250 cm


```python
runner = processor.Runner(
    executor=processor.IterativeExecutor(),
    #executor=processor.FuturesExecutor(),
    schema=ffschema.FFSchema,
    #maxchunks=2,
    )

p = sidm_processor.SidmProcessor(
    [
        "baseNoLj",
        #"baseNoLj_promptA",
        #"baseNoLj_displacedA",
        #"baseNoLj_veryDisplacedA",
        "baseNoLj_lowLxyA",
        "baseNoLj_highPtA",
        "baseNoLj_Lxy_upto250A"
        #"2mu2e",
        
        
    ],
    ["gen_base"]
)

output = runner.run(fileset, treename="ffNtuplizer/ffNtuple", processor_instance=p)
```

```python
ch1= "baseNoLj"
ch2= "baseNoLj_lowLxyA"
ch3= "baseNoLj_highPtA"
ch4= "baseNoLj_Lxy_upto250A"
```

## Pt distribution with Lxy cut

#### Here I show the LJ-reco efficiency w.r.t  Dark Photon PT plot with and with out the Lxy cut for the 0.25 GeV dark Photon for the two categories: One which decays to muons and one which decays to electron.
