```python
ch1 = "baseNoLj_A_ee_highPtA_lowLxyA"
ch2 = "baseNoLj_A_ee_matched_lj_highPT_lowLxyA"
```

```python
histograms = [
              ["genAs_toE_lxy_EEdR_XXLowRange",
              "genAs_toE_pt_EEdR_XXLowRange",
              "genAs_toE_lxy_pt_lowRange"
              ],
              ["genAs_toE_lxy_EEdR_XLowRange",
              "genAs_toE_pt_EEdR_XLowRange",
              "genAs_toE_lxy_pt_lowRange"
              ],
              ["genAs_toE_lxy_EEdR",
              "genAs_toE_pt_EEdR_lowRange",
              "genAs_toE_lxy_pt_lowRange"
              ]
             ]
     
hist_x_axis = [r"$Z_d$ $L_{xy}$", 
               r"$Z_d$ $p_T$", 
               r"$Z_d$ $L_{xy}$"
              ]
hist_y_axis = [r"$\Delta$R(e, e)", 
               r"$\Delta$R(e, e)",  
               r"$Z_d$ $p_T$"
              ]
hist_x_name = ["genA_lxy", 
               "genA_pt", 
               "genA_lxy",
              ]
hist_y_name = ["genE_genE_dR_lowRange",
               "genE_genE_dR_lowRange",
               "genA_pt",
              ]
```

```python
row = len(samples)
column = len(histograms)
fig , ax = plt.subplots(row, column, figsize =(row *12, column*10)) 
for s in range(row):
    for h in range(column):
        values = output["out"][samples[s]]["hists"][histograms[s][h]][ch2,:, :].values()\
          /(output["out"][samples[s]]["hists"][histograms[s][h]][ch1,:, :].\
            values() )
        hist2d = output["out"][samples[s]]["hists"][histograms[s][h]][ch1,:, :]
        mesh = ax[s,h].pcolormesh(*hist2d.axes.edges.T, values.T, vmin = 0.0 , vmax = 1.0)
        ax[s,h].set_xlabel(hist_x_axis[h])
        ax[s,h].set_ylabel(hist_y_axis[h])
        ax[s,h].set_title(samples[s])
        fig.colorbar(mesh)
        
```
